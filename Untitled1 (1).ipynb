```python
"""
E-Commerce Sales Analysis - Mini Project
Concepts: Data Cleaning | GroupBy | Sorting | Aggregation
"""

import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(42)


# CREATE RAW DATA (with intentional issues)

print("=" * 60)
print("E-COMMERCE SALES ANALYSIS")
print("=" * 60)

# Products catalog
products = ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones']
categories = ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Accessories']
prices = [899.99, 29.99, 79.99, 249.99, 59.99]

# Generate raw data with missing values and duplicates
data = {
    'order_id': [f'ORD-{i}' for i in range(1, 31)],
    'product': [np.random.choice(products) for _ in range(30)],
    'quantity': [np.random.randint(1, 6) for _ in range(30)],
    'price': [0]*30,  # Will fill below
    'customer': [f'CUST-{np.random.randint(1, 11)}' for _ in range(30)],
    'date': [
        datetime(2024, np.random.randint(1, 5), np.random.randint(1, 29))
        for _ in range(30)
    ],
    'status': ['Completed']*30
}

# Introduce data issues
data['price'][3] = -899.99        # Negative price
data['price'][7] = None           # Missing price
data['price'][12] = -29.99        # Negative price
data['quantity'][15] = 0          # Zero quantity
data['status'][8] = 'Cancelled'   # Cancelled order
data['status'][20] = 'Refunded'   # Refunded order
data['price'][25] = None          # Missing price

# Fill prices based on product
price_map = dict(zip(products, prices))
for i in range(30):
    if data['price'][i] == 0:
        data['price'][i] = price_map[data['product'][i]]

# Add duplicate rows
data['order_id'].append('ORD-5')
data['product'].append('Mouse')
data['quantity'].append(2)
data['price'].append(29.99)
data['customer'].append('CUST-3')
data['date'].append(datetime(2024, 2, 15))
data['status'].append('Completed')

data['order_id'].append('ORD-18')
data['product'].append('Monitor')
data['quantity'].append(1)
data['price'].append(249.99)
data['customer'].append('CUST-7')
data['date'].append(datetime(2024, 3, 10))
data['status'].append('Completed')

# Create DataFrame
df = pd.DataFrame(data)
df['total'] = df['quantity'] * df['price']

print(f"\nRAW DATA: {len(df)} rows\n")
print(df[['order_id', 'product', 'quantity', 'price', 'total', 'customer', 'status']].head(10).to_string())


# 1. DATA CLEANING

print("\n" + "=" * 60)
print("1. DATA CLEANING")
print("=" * 60)

# Remove duplicates
before = len(df)
df = df.drop_duplicates(subset='order_id')
print(f"Duplicates removed: {before - len(df)}")

# Handle missing values
print(f"Missing prices before: {df['price'].isna().sum()}")
df['price'] = df.groupby('product')['price'].transform(lambda x: x.fillna(x.median()))
print(f"Missing prices after:  {df['price'].isna().sum()}")

# Fix negative prices (take absolute value)
neg_count = (df['price'] < 0).sum()
df['price'] = df['price'].abs()
print(f"Negative prices fixed: {neg_count}")

# Remove invalid rows
before = len(df)
df = df[df['quantity'] > 0]
print(f"Zero-quantity rows removed: {before - len(df)}")

# Filter to valid orders only
before = len(df)
df = df[df['status'] == 'Completed']
print(f"Non-completed orders removed: {before - len(df)}")

# Recalculate total after cleaning
df['total'] = (df['quantity'] * df['price']).round(2)

# Add derived columns
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month_name()
df['category'] = df['product'].map(dict(zip(products, categories)))

print(f"\nCLEAN DATA: {len(df)} rows, {df['order_id'].nunique()} unique orders")
print(df[['order_id', 'product', 'quantity', 'price', 'total', 'customer', 'month']].head(8).to_string())


# 2. GROUPBY + AGGREGATION

print("\n" + "=" * 60)
print("2. GROUPBY + AGGREGATION")
print("=" * 60)

# Product-level aggregation
print("\n--- Product Performance ---")
product_stats = df.groupby(['product', 'category']).agg(
    total_revenue=('total', 'sum'),
    avg_price=('price', 'mean'),
    total_units=('quantity', 'sum'),
    order_count=('order_id', 'count'),
    avg_order_value=('total', 'mean'),
    min_order=('total', 'min'),
    max_order=('total', 'max')
).round(2)
print(product_stats.to_string())

# Customer-level aggregation
print("\n--- Customer Purchases ---")
customer_stats = df.groupby('customer').agg(
    total_spent=('total', 'sum'),
    total_orders=('order_id', 'nunique'),
    total_items=('quantity', 'sum'),
    avg_spent=('total', 'mean'),
    favorite_product=('product', lambda x: x.mode().iloc[0] if not x.mode().empty else 'N/A')
).round(2)
print(customer_stats.to_string())

# Monthly aggregation
print("\n--- Monthly Sales ---")
monthly_stats = df.groupby('month').agg(
    revenue=('total', 'sum'),
    orders=('order_id', 'nunique'),
    items_sold=('quantity', 'sum'),
    unique_customers=('customer', 'nunique'),
    avg_revenue_per_order=('total', 'mean')
).round(2)
print(monthly_stats.to_string())

# Category aggregation
print("\n--- Category Summary ---")
cat_stats = df.groupby('category').agg(
    revenue=('total', 'sum'),
    units_sold=('quantity', 'sum'),
    products=('product', 'nunique'),
    orders=('order_id', 'nunique')
).round(2)
print(cat_stats.to_string())

# Multi-level GroupBy: Category + Product
print("\n--- Category x Product Breakdown ---")
cat_prod = df.groupby(['category', 'product']).agg(
    revenue=('total', 'sum'),
    units=('quantity', 'sum'),
    orders=('order_id', 'count')
).round(2)
print(cat_prod.to_string())

# 3. SORTING
print("\n" + "=" * 60)
print("3. SORTING")
print("=" * 60)

# Top products by revenue (descending)
print("\n--- Top Products by Revenue (High to Low) ---")
top_products = product_stats.sort_values('total_revenue', ascending=False)
print(top_products[['total_revenue', 'total_units', 'order_count']].to_string())

# Bottom products by revenue (ascending)
print("\n--- Bottom Products by Revenue (Low to High) ---")
bottom_products = product_stats.sort_values('total_revenue', ascending=True)
print(bottom_products[['total_revenue', 'total_units', 'order_count']].to_string())

# Top customers by spend (descending)
print("\n--- Top Customers by Spend (High to Low) ---")
top_customers = customer_stats.sort_values('total_spent', ascending=False)
print(top_customers[['total_spent', 'total_orders', 'total_items']].head(8).to_string())

# Monthly revenue sorted chronologically (by month order)
month_order = ['January', 'February', 'March', 'April']
monthly_stats_sorted = monthly_stats.reindex(month_order)
print("\n--- Monthly Revenue (Chronological) ---")
print(monthly_stats_sorted[['revenue', 'orders', 'items_sold']].to_string())

# Orders sorted by total value (descending)
print("\n--- Individual Orders by Value (High to Low) ---")
top_orders = df.sort_values('total', ascending=False)
print(top_orders[['order_id', 'customer', 'product', 'quantity', 'total']].head(10).to_string(index=False))

# Multi-level sort: category (A-Z), then revenue (descending)
print("\n--- Products Sorted by Category (A-Z), then Revenue (High to Low) ---")
cat_alpha = product_stats.reset_index().sort_values(
    ['category', 'total_revenue'], ascending=[True, False]
)
print(cat_alpha[['category', 'product', 'total_revenue', 'total_units']].to_string(index=False))

# 4. SUMMARY REPORT
print("\n" + "=" * 60)
print("4. SUMMARY REPORT")
print("=" * 60)

total_rev = df['total'].sum()
total_orders = df['order_id'].nunique()
total_items = df['quantity'].sum()
total_customers = df['customer'].nunique()

print(f"\nOVERALL METRICS:")
print(f"  Total Revenue:     ${total_rev:>8,.2f}")
print(f"  Total Orders:      {total_orders:>8}")
print(f"  Total Items Sold:  {total_items:>8}")
print(f"  Unique Customers:  {total_customers:>8}")
print(f"  Avg Order Value:   ${total_rev/total_orders:>8,.2f}")
print(f"  Items per Order:   {total_items/total_orders:>8.2f}")

print(f"\nBEST PERFORMERS:")
best_prod = top_products.index[0]
best_cust = top_customers.index[0]
print(f"  Best Product:      {best_prod} (${product_stats.loc[best_prod, 'total_revenue']:.2f})")
print(f"  Best Customer:     {best_cust} (${customer_stats.loc[best_cust, 'total_spent']:.2f})")
print(f"  Best Month:        {monthly_stats['revenue'].idxmax()} (${monthly_stats['revenue'].max():.2f})")

print(f"\nDATA QUALITY:")
print(f"  Initial rows:      {len(data['order_id'])}")
print(f"  Final clean rows:  {len(df)}")
print(f"  Records removed:   {len(data['order_id']) - len(df)}")
print(f"  Data loss:         {((len(data['order_id']) - len(df)) / len(data['order_id']) * 100):.1f}%")

print("\n" + "=" * 60)
print("ANALYSIS COMPLETE")
print("=" * 60)
```
